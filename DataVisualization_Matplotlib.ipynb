```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
```

```python
dt=pd.read_csv('data/csv_data_1.csv')
dt
```

```python
dt.fillna(0, inplace=True)
```

```python
pyplot.plot(dt['id'],dt['sales'],'--y')  # parameters: x,y, formatting & colour
```

```python
pyplot.plot(dt['id'],dt['sales'],'.r',markersize=25)
```

```python
pyplot.plot(dt['id'],dt['sales'],'+g',markersize=25)
```

```python
pyplot.plot(dt['id'],dt['sales'],'^g',markersize=25)
```

```python
pyplot.plot(dt['id'],dt['sales'],'*b',markersize=25)
pyplot.title('Product Sales')
pyplot.xlabel('Product ID')
pyplot.ylabel('Sales')
```

## Other types of plots with Matplotlib

```python
pyplot.scatter(dt['id'],dt['sales'],color='orange',linewidth=30)
```

```python
dt['sales'].mean()
```

```python
pyplot.barh(dt['id'],dt['sales'])
```

```python
pyplot.barh(dt['id'],[83.125,83.125,83.125,83.125,83.125,83.125,83.125,83.125],label='Avg. Sales')
pyplot.barh(dt['id'],dt['sales'],label='Act. Sales')
pyplot.legend()
```

```python
pyplot.pie(dt['sales'],labels=dt['name'],radius=2)
pyplot.legend()
```
